```python
# Organizando uma lista de forma permanente utilizando o método sort().


cars = ['bmw', 'audi', 'toyota', 'subaru']
cars.sort()
print(cars)
#['audi', 'bmw', 'subaru', 'toyota']
# A lista foi organizada em ordem alfabética.
# Passando o argumento reverse=True, ordenamos  a lista de forma inversa. EX:

cars.sort(reverse=True)
print(cars)
# ['toyota', 'subaru', 'bmw', 'audi']

# Ordenando a lista temporariamente usando a função sorted(). Repare que aqui é uma função()
# e não método(). A função sorted() fica dentro da função print(sorted(cars)).

cars = ['bmw', 'audi', 'toyota', 'subaru']
print("\n\tHere is the original list:")
print(cars)
print("\n\tHere is the sorted list:")
print(sorted(cars))
# ['audi', 'bmw', 'subaru', 'toyota']
#Exibindo uma lista em ordem inversa "sem reorganizar a ordem alfabética" usando o método reverse().
cars.reverse()
print(cars)
#['subaru', 'toyota', 'audi', 'bmw']

# Descobrindo o tamanho de uma lista usando a função len().
print(len(cars))
# 4
# Nota: Python conta os itens de uma lista começando em um, portanto você não deverá se deparar com
# nenhum erro de deslocamento de um ao determinar o tamanho de uma lista. Já o índice de uma lista,
# começa em zero [0].

# Evitando erros de índice. No exemplo abaixo, ao digitar 3 estarei buscando o quarto elemento que
# não existe. Pois em python o índice começa em zero [0]. No exemplo abaixo seria exibido um erro de índice.

motorcycles = ['honda', 'yamaha', 'suzuki']
print(motorcycles[2])
# print(motorcycles[3])
# Traceback (most recent call last):
# File "C:\Users\marce\PycharmProjects\estudo\livro_cap_03_car.py", line 39, in <module>
# print(motorcycles[3])
# IndexError: list index out of range. Caso quisesse  selecionar "suzuki" deveria digitar [2].
# suzuki
```

```python
# 4 - TRABALHANDO COM LISTAS

#  Quando quiser executar a mesma ação em todos os itens de uma lista, você pode usar o
#  laço " for " de Python.

magicos = ["alice", "dualipa", "britney", "madonna\n"]
for magico in magicos:
    print(magico)
#alice
#dualipa
#britney
#madonna

# Na linha "for" Python diz  para extrair um nome da lista magicos e armazená-lo na variável magico.
# Em seguida dizemos a Python  para exibir o nome que acabou de ser armazenado em magico.
# A leitura da linha de código seria:  “para todo mágico na lista de mágicos, exiba o nome do mágico”

# É conveniente escolher um nome significativo, que represente um único item da lista. Por exemplo,
# eis uma boa maneira de iniciar um laço " for " para uma lista de gatos, uma lista de cachorros e uma
# lista genérica de itens:

# for cat in cats:
# for dog in dogs:
# for item in items:

magicos = ["alice", "dualipa", "britney", "madonna"]
for magico in magicos:
    print(magico.title() + ", Isso foi um belo truque.")
    print("Mal posso esperar para ver o próximo.\n")
print("********** Até a próxima!!! ********** ")

# Qualquer linha de código após o laço "for" que não estiver indentada será executada
# uma vez, sem repetição como no exemplo acima " Até a próxima ".

#EVITANDO ERROS DE IDENTAÇÃO.
# Sempre indente a linha após a instrução "for" em um laço. Se você se esquecer, Python o avisará; Ex:

magicians = ['\nalice', 'david', 'carolina']
for magician in magicians:
    print(magician)
#   File "C:\Users\marce\PycharmProjects\estudo\livro_cap_04.py", line 40
#     print(magician)
#     ^
# IndentationError: expected an indented block

# Esquecendo-se de indentar linhas adicionais
# Às vezes, seu laço executará sem erros, mas não produzirá o resultado esperado. Isso pode acontecer
# quando você tenta realizar várias tarefas em um laço e se esquece de indentar algumas de suas
# linhas.

#Indentando Desnecessariamente:
# Se você, acidentalmente, indentar uma linha que não precisa ser indentada, Python o informará a
# respeito da indentação inesperada.
#  Como Python não sabe o que você está querendo fazer com seu código, ele executará todo código
#  que estiver escrito com uma sintaxe válida. Se uma ação for repetida muitas vezes quando deveria
#  ser executada apenas uma vez, verifique se você não precisa simplesmente deixar de indentar
#  o código dessa ação.

# Esquecendo os dois-pontos:
# Os dois-pontos no final de uma instrução "for" diz a Python para interpretar a próxima linha como
# o início de um laço. Ex:

magicians = ['\nalice', 'david', 'carolina']
for magician in magicians:  # está faltando os : dois pontos no final do laço "for".
    print(magician)
# File "C:\Users\marce\PycharmProjects\estudo\livro_cap_04.py", line 64
#     for magician in magicians
#                              ^
# SyntaxError: invalid syntax
```

```python
# CRIANDO LISTAS NUMÉRICAS:

# Usando a função range()
# A função range() de Python facilita gerar uma série de números. Por exemplo, podemos usar a
# função range() para exibir uma sequência de números, assim:

for value in range(1,6):
    print(value)
#1 2 3 4 5 . Embora esse código dê a impressão de que deveria exibir os números de 1 a 6, ele não
# exibe o número 6. O Python sempre desloca em 1 o último valor. Para exibir o 6, deveria colocar (1,7)

# Para criar listas com função range () podemos usar list() para converter esse mesmo conjunto
# de números em uma lista. Ex:
numbers = list(range(1,6))
print(numbers)
# [1, 2, 3, 4, 5]

#Listando números pares de 0 a 10:
numeros_pares = list(range(0,11,2))
print(numeros_pares)
# [0, 2, 4, 6, 8, 10]

# Criando uma lista dos dez primeiros quadrados perfeitos (isto é, o quadrado de cada
# inteiro de 1 a 10). Ex:

quadrados = []
for value in range(1,11):
    quadrado = value**2
    quadrados.append(quadrado)
print(quadrados)
# [1, 4, 9, 16, 25, 36, 49, 64, 81, 100]
# O mesmo código de um modo mais conciso. Lista com o quadrado dos números
# ímpares de 1 a 11 .Ex:

quadrados = []
for value in range(1,12,2):
    quadrados.append(value**2)
print(quadrados)
# [1, 9, 25, 49, 81, 121]

# Obs:  Às vezes, usar uma variável temporária deixa o código mais legível; em outras ocasiões, deixa
#  o código desnecessariamente longo. Concentre-se primeiro em escrever um código que você
#  entenda claramente e faça o que você quer que ele faça. Em seguida, procure abordagens mais
#  eficientes à medida que revisar seu código.

# Estatísticas Simples Com uma Lista de Números:
# Algumas funções Python são específicas para listas de números. Por exemplo, podemos encontrar
# facilmente o valor mínimo, o valor máximo e a soma de uma lista de números:

numeros = [1,2,3,4,5,6,7,8,9]
print(min(numeros))
print(max(numeros))
print(sum(numeros))
# 1
# 9
# 45
# List comprehensions:
# A abordagem descrita antes para gerar a lista "quadrados" usou três ou quatro linhas de código. Uma
# list comprehension (abrangência de lista) permite gerar essa mesma lista com apenas uma linha
# de código. A lista com o quadrado dos números ímpares gerada anteriormente será construída
# agora com somente uma linha de código.

quadrados = [value**2 for value in range(1,12,2)]
print(quadrados)
# [1, 9, 25, 49, 81, 121]
# observe que desta vez não colocamos os dois-pontos no final da instrução "for", pois está inserida dentro
# dos conchetes.

```

```python
# TUPLAS

# As listas funcionam bem para armazenar conjuntos de itens que podem mudar durante
# a vida de um programa. No entanto, às vezes, você vai querer criar uma lista de itens que
# não poderá mudar. As tuplas permitem fazer exatamente isso. Python refere-se a valores que
# não podem mudar como imutáveis, e uma lista imutável é chamada de tupla.

# Uma tupla se parece exatamente com uma lista, exceto por usar parênteses no lugar de
# colchetes. Depois de definir uma tupla, podemos acessar elementos individuais usando o
# índice de cada item, como faríamos com uma lista. Ex:

dimensions = (10, 200, 300)
print(dimensions[0])
print(dimensions[1])
# 10
# 200

for dimension in dimensions:
    print(dimension)

# 10
# 200
# 300

# Sobrescrevendo uma tupla:
# Embora não seja possível modificar uma tupla, podemos atribuir um novo valor a uma variável
# que armazena uma tupla. Portanto, se quiséssemos alterar nossas dimensões, poderíamos redefinir
# a tupla toda:

dimensions = (200, 50)
print("\nOriginal dimensions:")
for dimension in dimensions:
    print(dimension)

dimensions = (400, 100)
print("\n Modified dimensions: ")
for dimension in dimensions:
    print(dimension)

# Se comparada com listas, as tuplas são estruturas de dados simples. Use-as quando quiser
# armazenar um conjunto de valores que não deva ser alterado durante a vida de um programa.
```

```python
# INSTRUÇÕES IF

cars = ['audi', 'BmW', 'subaru', 'toyota']
for car in cars:
    if car.lower() == "bmw":
        print(car.upper())
    else:
        print(car.title())
# Audi
# BMW
# Subaru
# Toyota
# Um único sinal de igual, na verdade, é uma instrução; você poderia ler o código como “defina o valor de car
# como 'audi'”. Por outro lado, um sinal de igualdade duplo, faz uma pergunta: “O valor de car é igual a 'bmw'?”.
# A maioria das linguagens de programação utiliza sinais de igualdade dessa maneira.

# Testes de igualdade diferenciam letras maiúsculas de minúsculas em Python. Por exemplo, dois valores com
# diferenças apenas quanto a letras maiúsculas ou minúsculas não são considerados iguais:
# car = 'Audi'
# car == 'audi'
# False

# Porém, se essa diferença não importar e você simplesmente quiser testar o valor de uma variável, poderá
# converter esse valor para letras minúsculas antes de fazer a comparação. Como exemplo acima:
# car = "BmW"
# car.lower() == "bmw"
# True

# Verificando a diferença:
# Se quiser verificar se dois valores não são iguais, você poderá combinar um ponto de exclamação e um
# sinal de igualdade (!=). O ponto de exclamação representa não, como ocorre em muitas linguagens de
# programação. Ex:

ingredientes_pedidos = "musarela"
if ingredientes_pedidos != "calabresa":
    print("\nMantenha a calabresa.")
else:
    print("\nNão coloque calabresa.")
# Mantenha a calabresa

ingredientes_pedidos = "calabresa"
if ingredientes_pedidos != "calabresa":
    print("\nMantenha a calabresa.")
else:
    print("\nNão coloque calabresa.")
# Não coloque calabresa.

# Observação: A maioria das expressões condicionais que você escrever testará a igualdade, porém, às
# vezes, achará mais eficiente testar a não igualdade.

# Comparações Numéricas:
# Testando várias condições : pode-se usar "and" ou "or"
# u	>>> age_0 = 22
# >>> age_1 = 18
#  v >>> age_0 >= 21 and age_1 >= 21
# False
# w	>>> age_1 = 22
# v >>> age_0 >= 21 or age_1 >= 21
# True
# >>> age_0 >= 21 and age_1 >= 21
# True

# Para verificar se duas condições são True simultaneamente, utilize a palavra reservada "and" para combinar os
# dois testes condicionais. Já a palavra reservada "or" permite verificar várias condições também, mas o teste
# passa se um dos testes individuais passar, ou ambos. Uma expressão "or" falha somente quando os dois
# testes individuais falharem.

# Verificando se um valor não está em uma lista


usuarios_banidos = ['andrew', 'carolina', 'david']
usuario = 'CAROLINA'

if usuario.lower() not in usuarios_banidos:
    print("\n"+usuario.title() + ", you can post a response if you wish.")
else:
    print("\n"+usuario.title() + ", você está temporiamente proibido de postar.")

# Carolina, você está temporiamente proibido de postar.
```

```python
# Testando Várias condições

# Quando for importante verificar todas as condições de interesse, deve-se usar uma série de
# instruções "if" simples, sem blocos "elif" ou "else". Essa técnica faz sentido quando mais de uma condição
# pode ser "True" e você quer atuar em todas as condições que sejam verdadeiras. Exemplo da pizzaria:

ingredientes_pedidos = ['pepperoni', 'queijo', 'calabresa']

if 'pepperoni' in ingredientes_pedidos:
    print("\nAdicione pepperone.")

if 'calabresa' in ingredientes_pedidos:
    print("\nAdicione calabresa.")

if 'queijo' in ingredientes_pedidos:
    print("\nAdicione queijo.")

print("\nFinished!! Making your pizza!\n")

# Em suma, se quiser que apenas um bloco de código seja executado, utilize uma cadeia if-elif-else. Se for mais
# de um bloco de código a executar, utilize uma série de instruções "if" independentes.

# Usando Instruções If com Listas:

ingredientes = ['calabresa', 'pimenta malagueta', 'queixo']

for ingrediente in ingredientes:
    print("Adicione " + ingrediente + ".")
print("\n A pizza está pronta!!.\n")

# Caso não tenha um ingrediente:
ingredientes = ['calabresa', 'pimenta malagueta', 'queixo']

for ingrediente in ingredientes:
    if ingrediente == "pimenta malagueta":
        print("Desculpe!! Estamos sem pimenta malagueta.")
    else:
        print("Adicione " + ingrediente + ".")
print("\n A pizza está pronta!!.")

# VERIFICANDO SE UMA LISTA NÃO ESTÁ VAZIA

ingredientes = []

if ingredientes:
    for ingrediente in ingredientes:
        print("\nAdicione " + ingrediente + ".")
    print("\nA pizza está pronta!!.\n")
else:
    print("\nEstamos sem pizza no momento.\n")
#  Estamos sem pizza no momento.

# USANDO VÁRIAS LISTAS:

ingredientes_disponiveis = ['mushrooms', 'olives', 'green peppers', 'pepperoni', 'pineapple', 'extra cheese']

pedido_dos_clientes = ['mushrooms', 'french fries', 'extra cheese']

for pedido_do_cliente in pedido_dos_clientes:
 if pedido_do_cliente in ingredientes_disponiveis:
     print("Adding " + pedido_do_cliente + ".")
 else:
     print("Sorry, we don't have " + pedido_do_cliente + ".")

print("\nFinished making your pizza!")
```
